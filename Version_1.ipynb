```python
from rembg import remove
from PIL import Image
```

```python
input_file_name = "resize.jpg"
processed_image_file= "resize_only.png"
```

```python
in_file = Image.open(input_file_name)
remove_background = remove(in_file)
remove_background.save(processed_image_file)
```

```python
from IPython.display import Image 
input_img = Image(filename=input_file_name)
display(input_img)
```

```python
final_img = Image(filename=processed_image_file)
display(final_img)
```
